```python
t = LabelEncoder()
```

```python
t.fit_transform(df.sex)
```

```python
df['sex'] = t.fit_transform(df.sex)
```

```python
df['region'] = t.fit_transform(df.region)
```

```python
df
```

```python
y = df.charges 
x = df.iloc[:,:-1]
```

```python
y = y.values
x = x.values
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x, y)
```

```python
model.predict(x)
```
